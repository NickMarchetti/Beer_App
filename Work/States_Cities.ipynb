```python
mapq_url = "http://open.mapquestapi.com/geocoding/v1/reverse?key=KEY&location=30.333472,-81.470448&includeRoadMetadata=true&includeNearestIntersection=true"

state_list = []
city_list = []

for i in range(len(beer_lats)):
#for i in range(5):
    coordinates = str(beer_lats[i]) + "," + str(beer_lngs[i])
    
    base_url = "http://open.mapquestapi.com/geocoding/v1/reverse?key=" + mq_key + "&location=" + coordinates + "&includeRoadMetadata=true&includeNearestIntersection=true"

    response = requests.get(base_url)
    data = (response.json())
    
    city = data['results'][0]['locations'][0]['adminArea5']
    state = data['results'][0]['locations'][0]['adminArea3']
    
    try:
        city_list.append(city)
        state_list.append(state)
        
        print(city + " " + state)
        
    except:
        city_list.append("Not Found")
        state_list.append("Not Found")
        
        print("Not Found")
```

```python
beer_locations_df['City'] = city_list
beer_locations_df['State'] = state_list

beer_locations = beer_locations_df.sort_values(by=['Brewery']).reset_index(drop = True)

del beer_locations['Unnamed: 0']
beer_locations
```
