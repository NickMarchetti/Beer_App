```python
state_list = []
state_abbr_list = []
city_list = []

for i in range(len(beer_lats)):
#for i in range(5):
    coordinates = str(beer_lats[i]) + "," + str(beer_lngs[i])
    
    base_url = "https://maps.googleapis.com/maps/api/geocode/json?latlng=" + coordinates + "&key=" + gkey

    response = requests.get(base_url)
    data = (response.json())
    address = data['results'][3]['address_components']
    
    for x in range(len((data['results'][3]['address_components']))):
        if address[x]['types'][0] == 'locality':
            city = address[x]['long_name']
            city_list.append(city)
        if address[x]['types'][0] == 'administrative_area_level_1':
                state = address[x]['long_name']
                state_list.append(state)
                state_abbr = address[x]['short_name']
                state_abbr_list.append(state_abbr)
            
        print(city + " | " + state + " | " + state_abbr)
                
    except:
        city_list.append("Not Found")
        state_list.append("Not Found")
        state_abbr_list.append("Not Found")
```
